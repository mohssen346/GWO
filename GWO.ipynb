```python
import numpy as np
import numpy as np
import copy
```

```python
def sphere_fitness(X):
    fit_val = 0
    for i in range(len(X)):
        fit_val = fit_val + X[i]*X[i]
    return fit_val
```

```python
import numpy as np
class GWO:
    def __init__(self, fitness, dim, minx, maxx):
        
        self.position = [0 for i in range(dim)]
        
        for i in range(dim):
            self.position[i] = ((maxx-minx)* np.random.rand()+minx)
            
        self.fitness = sphere_fitness(self.position)
        
    def main(fitness, max_iter, n, dim, minx, maxx):
        
        population = [GWO(fitness, dim, minx, maxx) for i in range(n)]
        population = sorted(population, key = lambda i: i.fitness) # asc
        
        alpha_wolf, beta_wolf, gamma_wolf = copy.copy(population[:3])
        Iter = 0
        while Iter < max_iter:
            
            print('iter:\n', Iter, '\nbest_solution:', alpha_wolf.position,
                 '\nbest_fitness:', alpha_wolf.fitness ,'\n***************')
            
            a = 2*(1-Iter/max_iter) # 2----0
            
            for i in range(n):
                A1, A2, A3 = a*(2*np.random.rand()-1), a*(2*np.random.rand()-1), a*(2*np.random.rand()-1)
                C1, C2, C3 = 2*np.random.rand(), 2*np.random.rand(), 2*np.random.rand()
              
                X1 = [0 for i in range(dim)]
                X2 = [0 for i in range(dim)]
                X3 = [0 for i in range(dim)]
                Xnew = [0 for i in range(dim)]
                
                for j in range(dim):
                    X1[j] = alpha_wolf.position[j]-A1* abs(C1 * alpha_wolf.position[j] - population[i].position[j])
                    X2[j] = beta_wolf.position[j]-A2*abs(C2* beta_wolf.position[j] - population[i].position[j])
                    X3[j] = gamma_wolf.position[j]-A3*abs(C3* gamma_wolf.position[j] - population[i].position[j])
                    Xnew[j] = Xnew[j]+ X1[j]+X2[j]+X3[j]
                
                for j in range(dim):
                    Xnew[j] = Xnew[j]/3
                    
                fit_new = fitness(Xnew)
                
                if fit_new < population[i].fitness: #min
                    population[i].position = Xnew
                    population[i].fitness = fit_new
                
                population = sorted(population, key = lambda i: i.fitness) # asc
        
                alpha_wolf, beta_wolf, gamma_wolf = copy.copy(population[:3])
            
            Iter = Iter +1
         
        return alpha_wolf.position
```

```python
dim = 3
fitness = sphere_fitness
max_iter = 1000
num_wolf = 50
minx, maxx = -10,10
best_position = GWO.main(fitness, max_iter, num_wolf, dim, minx, maxx)
```
